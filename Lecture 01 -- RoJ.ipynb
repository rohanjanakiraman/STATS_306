```r
ggplot(data=filter(mpg, class!='2seater'), aes(x=displ, y=hwy)) + 
    geom_point(aes(color=class)) + geom_smooth() #You want to pass in the filtered data rather than the unfiltered data
```

```r
#Make sure to first filter your data
```

```r
ggplot(data=filter(mpg, class!='2seater'), aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth()
```

```r
#Where you add the color argument matters 
```

```r
mpg_filter = filter (mpg, displ < 5)
ggplot(mpg_filter, data=filter(mpg, class!='2seater'), aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth()
```

```r
mpg_filter = filter (mpg, displ < 5)
ggplot(mpg_filter, aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth()
```

```r
mpg_filter = filter (mpg, displ < 5)
ggplot(mpg_filter, aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth(data = filter(mpg, class == 'subcompact'))
```

```r
mpg_filter = filter (mpg, displ < 5, class!='suv')
ggplot(mpg_filter, aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth(data = filter(mpg, class == 'subcompact'))
```

```r
#== tests for equality
# = means assignment
```

```r
mpg_filter = filter (mpg, displ < 5, class!='suv')
ggplot(mpg_filter, aes(x=displ, y=hwy, shape = class)) + 
    geom_point() + geom_smooth(data = filter(mpg, class == 'compact'))
```

```r
mpg_filter = filter (mpg, displ < 5, manufacturer!='volkswagen') #Sorts out Volkswagen 
ggplot(mpg_filter, aes(x=displ, y=hwy, shape = class)) + 
    geom_point() + geom_smooth(data = filter(mpg, class == 'compact'))
```

```r
mpg_filter = filter (mpg, displ < 5, class!='suv')
ggplot(mpg_filter, aes(x=displ, y=hwy, color = class)) + 
    geom_point() + geom_smooth(data = filter(mpg, class == 'subcompact')) + geom_jitter() #Jitter adds random noise
```

```r
mpg_filter = filter (mpg, displ < 5, class!='suv')
ggplot(mpg_filter, aes(x=displ, y=hwy, color = manufacturer)) + 
    geom_point() + geom_smooth(data = filter(mpg, manufacturer == 'audi')) + geom_jitter()
```
